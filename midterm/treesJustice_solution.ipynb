##  Investigation of access to green as a function of socioeconomic status

The NYC trees are surveyed periodically in the tree census by the NYC parks department. The [latest trees census](https://www.nycgovparks.org/trees/treescount) (the third) occurred between 2015 and 2016. 

Studies show that access to vegetation has positive physical and psychological health effects on citizens in urban environments, but access to green is generally uneven in different areas of a city. Often more vulnerable sections of the city population (socioe-conomically disadvantaged areas and predominantly minority areas) have less access to natural resources. See for example [this study](
https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0122051)

This analysis is designed to test the hypothesis that ethnic diversity is correlated with access to green resources in NYC.

# Assess which of these distributions is most consistent with a Gaussian.

You used statistical tests that answer the question "can you reject the null hypothesis that a sample comes from a Standard Normal distriobution?"  

You do not need to frame this as a null hypothesis rejection test, but you can use the same test to compare how close to a Gaussian distribution each distribution is. 

Your deliverable for this portion is a test reault (or the results of a few tests) and its interpretation that shows which of the three distributions is closer to a Gaussian

What does the fact that some of these features are highly non Gaussian tell you?

** Note: since you are not asked to assess if they are statistically consistent with Gaussians, you do not need to set an alpha threshold and describe a null hypotehsis, you can use the value of the statistics to compare Gaussianity. note however that to commpare with a standard normal you have to standardize the distribution first!**

```python
print ("KS test between tree health by zipcode and a Gaussian distribution")
sp.stats.kstest((treepp.treehealth - treepp.treehealth.mean()) / 
                treepp.treehealth.std(), "norm")
```

```python
print ("KS test between number of trees per person, by zipcode, and a Gaussian distribution")
sp.stats.kstest((treepp.treepp - treepp.treepp.mean()) / treepp.treepp.std(), 
                "norm")
```

```python
print ("KS test between diversity (fraction of non-white population) and a Gaussian distribution")
sp.stats.kstest((treepp.non_whiteFraction - treepp.non_whiteFraction.mean()) / 
                treepp.non_whiteFraction.std(), "norm")
```

## conclusion: a small value of the KS statistics indicates a small difference between the cumulative distributions that are being compared. Thus the feature with smallest KS statistics, the tree health, is the most significantly Gaussian. A Gaussian distribution, in general, may suggest that the health of trees is generally random, i.e. unbiased, across the spatial areas considered. If we had seen for example a bimodal distribution, this may have been an indication of correlation between tree health and another feature, maybe diversity, but we do not. But more tests are certainly needed to assess it. This analysis is still at the exploratory level.

# Based on the result in 10, choose either tree health or trees per person and assess if the distribution of this variable is consistent with the population diversity.

This must be a test which returns a rejection (or non rejection) of a hypothesis. 



## with my data and my result for 10 either distribution can be tested, since the correlation is weak. 

** Note: To test if the "come from the same distribution" we should still standardize the features since they are not normalized quantities to begin with (one is a population fraction, the other an average of arbitrary values or a number of item per person**

** Threshold for significance: alpha = 0.1 **

** Null: the distribution of tree health and of diversity, as measured as fraction of non white population, come from the same parent distribution (and are statistically indistinguishable, p-value 0.1 **


```python
print("Running a KS test without standardizing the distributions (testing central tendency)")
sp.stats.ks_2samp(treepp.non_whiteFraction, treepp.treepp)
```

```python
print("Running a KS test after standardizing the distributions (testing shape)")
sp.stats.ks_2samp((treepp.non_whiteFraction - treepp.non_whiteFraction.mean()) / 
                treepp.non_whiteFraction.std(), 
                  (treepp.treepp - treepp.treepp.mean()) / treepp.treepp.std())
```

##  conclusions: the distributions of diversity and trees per person exhibit very different central tendencies (mean) so they are inherently different, but testing their standardize version allows us to see if the distribution *shape* is consistent. However, with a significant threshold of 0.1 we cannot rule out the Null hypothesis that the distribution come from the same parent distribution.
